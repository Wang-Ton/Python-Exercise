```python
import pandas as pd
import matplotlib.pyplot as plt
weeks = pd.read_excel(r'I:\Python\Weeks.xlsx',index_col='weeks')
#weeks.plot.area(y=['banana','pine','apple','strawberry'])
weeks.plot(y=['banana','pine','apple','strawberry'])
plt.title('Sales Weekly Trend',fontsize=16,fontweight='bold')
plt.xticks(weeks.index,fontsize=8)
plt.ylabel('Pieces',fontsize=12)
plt.show
```
