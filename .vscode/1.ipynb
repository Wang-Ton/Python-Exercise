```python
import pandas as pd
import matplotlib.pyplot as plt 
sells = pd.read_excel('I:\Python\sells.xlsx',index_col='name')
sells['2017'].sort_values(ascending=True).plot.pie(fontsize=8)
plt.title("sells counts",fontsize=12,fontweight='bold')
plt.ylabel('2017',fontsize=12,fontweight='bold')
plt.tight_layout()
plt.show
```

```python
import pandas as pd
import matplotlib.pyplot as plt 
sells = pd.read_excel('I:\Python\sells.xlsx',index_col='name')
#sells['2017'].sort_values(ascending=False).plot.pie(fontsize=8,startangle=-270)
sells['2017'].plot.pie(fontsize=8,counterclock=True,startangle=-270)
plt.title("sells counts",fontsize=12,fontweight='bold')
plt.ylabel('2017',fontsize=12,fontweight='bold')
f=plt.gcf()
f.subplots_adjust(left=0.2,bottom=0.42)
plt.show
```
